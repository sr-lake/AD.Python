# Exercícios 

1. Importe as bibliotecas necessárias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
y
```

2. Crie os dados e plote-os em um objeto axes. Ajuste os rótulos e o título do gráfico.

```python
plt.plot(x, y, 'r')
plt.xlabel('eixo X')
plt.ylabel('eixo Y')
plt.title('Gráfico Modelo')
plt.show()
```

```python

plt.subplot(1,2,1)
plt.title('Grafico - 01')
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.title('Grafico - 02')
plt.plot(y, x, 'g*-');
```

3. Crie um objeto do tipo figure e coloque dois eixo no mesmo, localizados em \[0, 0, 1, 1\] e \[0.2, 0.5, .2, .2\]. Então plote os mesmos dados em ambos.

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x,y)

```

4. Crie dois gráficos em um mesmo figure.

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # figura principal
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.25]) # inset

# Figura principal
axes1.plot(x, y, 'b')
axes1.set_xlabel('eixo X axes1')
axes1.set_ylabel('eixo Y axes1')
axes1.set_title('Título axes1')

# Figura do inset
axes2.plot(y, x, 'r')
axes2.set_xlabel('eixo X axes2')
axes2.set_ylabel('eixo Y axes2')
axes2.set_title('Título axes2');
```
